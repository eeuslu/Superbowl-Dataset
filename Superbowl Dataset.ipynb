### Import of necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

### Data Loading

```python
nfl_data = pd.read_csv("superbowl.csv") #read the .csv dataset
nfl_data.shape #present dimensions of the data
```

```python
nfl_data.head(5) #present first 5 columns of the data set
```

### Data information
&bull; **Date** - Month and Year of the Game

&bull; **SB** - SB is the Super Bowl Number

&bull; **Winner** - Winning Team of the Year

&bull; **Winner Pts** - Total Points of the winning team

&bull; **Loser** - Loser Team of the Year

&bull; **Loser Pts** - Total Points of the losing team

&bull; **MVP** - MVPis most valuable player of the match

&bull; **Stadium** - Stadium who host the game

&bull; **City** - City of the stadium

### Data Visualization

#### Comparison of Team vs. Team for all seasons
Plot presents how many times teams faced each other

```python
nfl_data['Win_vs_Lost'] = nfl_data['Winner'] + " vs." + nfl_data["Loser"]
plt.figure(figsize=(15,15))

sns.countplot(y="Win_vs_Lost",data=nfl_data,orient='h',order=nfl_data['Win_vs_Lost'].value_counts().sort_values(ascending=False).index)
sns.set_style("whitegrid")

plt.title("Team vs. Team comparison")
plt.xlabel("Number of Time Teams played each other")
plt.ylabel("NFL Teams")
plt.show()
```

teams faced each other just once once

#### Comparison of Team vs. Team in History by Years
Plot presents how many times teams faced each other with the outcome

```python
plt.figure(figsize=(15,15))
plt.xticks(rotation=75)

sns.scatterplot(x=nfl_data.index,y=nfl_data['Win_vs_Lost'],hue=nfl_data['Winner'],s=150)
sns.set_style("whitegrid")

plt.title("Number of Time Faced and Winers By Year") 
plt.ylabel("NFL Teams") 
plt.xlabel("Game Years")
```

#### Plotting all Winers
Plot presents what happened when each team faced other teams

```python
plt.figure(figsize=(10,0))
plt.xticks( rotation=90)

sns.scatterplot(x=nfl_data['Winner'],y=nfl_data['Loser'],hue=nfl_data['Winner'],s=200)
sns.set_style("whitegrid")


plt.title("Winners vs Losers Teams") 
plt.ylabel("Losers") 
plt.xlabel("Winners") 
```

#### Number of Winnings of NFL Super Bowl
Plot presents how many times a team won the NFL Super Bowl

```python
plt.figure(figsize=(10,10))
plt.xticks( rotation=75)

sns.countplot(x="Winner",data=nfl_data,order=nfl_data['Winner'].value_counts().sort_values(ascending=False).index)
sns.set_style("whitegrid")


plt.title("Wins by each Team") 
plt.ylabel("# of Wins") 
plt.xlabel("NFL Team") 
plt.show()
```

#### Number of Losts of NFL Super Bowl
Plot presents how many times a team lost the NFL Super Bow

```python
plt.figure(figsize=(10,10))
plt.xticks( rotation=75)

sns.countplot(x="Loser",data=nfl_data,order=nfl_data['Loser'].value_counts().sort_values(ascending=False).index)
sns.set_style("whitegrid")


plt.title("Loses by each Team") 
plt.ylabel("# of Loses") 
plt.xlabel("NFL Team") 
plt.show()
```

#### Winnings of Each Team of NFL Super Bowl by Mean Points

```python
plt.figure(figsize=(20,10))
plt.xticks( rotation=75)

sns.barplot(x=nfl_data['Winner'],y=nfl_data['Winner Pts'])
sns.set_style("whitegrid")


plt.title("Wins NFL Teams by Points") 
plt.ylabel("Points") 
plt.xlabel("NFL Team")
```

#### Loses of Each Team of NFL Super Bowl by Mean Points

```python
plt.figure(figsize=(20,10))
plt.xticks( rotation=75)

sns.barplot(x=nfl_data['Loser'],y=nfl_data['Loser Pts'])
sns.set_style("whitegrid")


plt.title("Loses NFL Teams by Points") 
plt.ylabel("Points") 
plt.xlabel("NFL Teams") 
```

#### Density of Winning and Losing Teams by Points

```python
plt.figure(figsize=(20,10))
plt.xticks( rotation=75)
sns.set_style("white")

sns.distplot(a=nfl_data['Winner Pts'], label="Winners")
sns.distplot(a=nfl_data['Loser Pts'], label="Loser")

plt.title("Histogram of Petal Lengths, by Species")

plt.legend()
```

#### Number of Most Valueble Player by Counts of Nominations

```python
plt.figure(figsize=(20,10))
plt.xticks( rotation=75)

sns.countplot(x="MVP",data=nfl_data,order=nfl_data['MVP'].value_counts().sort_values(ascending=False).index,palette="rocket")
sns.set_style("whitegrid")


plt.title("MVP of Game Nominated") 
plt.ylabel("# of Nominations") 
plt.xlabel("NFL Player") 
```

#### Number of Hosted Matches by Stadium Counts

```python
plt.figure(figsize=(20,10))
plt.xticks( rotation=85)

sns.countplot(x="Stadium",data=nfl_data,order=nfl_data['Stadium'].value_counts().sort_values(ascending=False).index,palette="rocket")

sns.set_style("whitegrid")


plt.title("Games played in most stadiums") 
plt.ylabel("# of Playes in Stadiums") 
plt.xlabel("Stadium") 
plt.show()
```
